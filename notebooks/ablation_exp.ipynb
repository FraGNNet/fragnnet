```python
import pandas as pd
import numpy as np
import os
```

```python
sims_df_fp = "../figs/ablation_exp/test_df.csv"
sims_latex_fp = "../figs/ablation_exp/test_df.tex"
sims_df = pd.read_csv(sims_df_fp)
display(sims_df)
```

```python
model_type_to_name = {
    "fraggnn_d3_noce": "FraGNNet-D3 (-CE)",
    "fraggnn_d3_edges": "FraGNNet-D3 (+Edges)",
    "fraggnn_d4_noce": "FraGNNet-D4 (-CE)",
    "fraggnn_d4_edges": "FraGNNet-D4 (+Edges)",
}

latex_df = sims_df[["model_type","cos_sim_0.01_mean","cos_hun_mean","cos_sim_0.01_std","cos_hun_std"]].copy()

#  latex_df = latex_df.drop(columns=["pred_oos_prob_mean","pred_oos_prob_std","pearson_r_oos_mean","pearson_r_oos_std"])
for metric in ["cos_sim_0.01","cos_hun"]:
    def fmt_func(row):
        mean_score = row[f"{metric}_mean"]
        std_score = row[f"{metric}_std"]
        if np.isnan(mean_score) or np.isnan(std_score):
            fmt_str = "-"
        else:
            fmt_str = f"${mean_score:.3f} \pm {std_score:.3f}$"
        return fmt_str
    latex_df[metric] = latex_df.apply(fmt_func, axis=1) 
    latex_df = latex_df.drop(columns=[f"{metric}_mean",f"{metric}_std"])
latex_df.loc[:,"model_type"] = latex_df["model_type"].replace(model_type_to_name)
# rename columns
latex_df.columns = ["Model",r"$\mathbb{C}_\text{BIN} \Uparrow$",r"$\mathbb{C}_\text{HUN} \Uparrow$"]
# latex_table = latex_df.to_latex(index=False, escape=False)
```
